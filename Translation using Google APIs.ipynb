```python
import pandas as pd
from google.cloud import translate_v2 as translate
import langdetect
import os
import time
import numpy as np

# Set up Google Cloud authentication
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "/Users/sreeharsha/Documents/TGH Data Management Cleaning/nth-infusion-452016-j7-9b0ca2e90993.json"

# Initialize Google Translate API client
client = translate.Client()

# Function to detect language
def detect_language(text):
    try:
        return langdetect.detect(text)
    except:
        return "unknown"  # If detection fails

# Function to clean text before translation
def clean_text(text):
    """Removes NaN, trims spaces, and filters out non-text data."""
    if pd.isna(text) or text is None:
        return ""  # Convert NaN to an empty string
    text = str(text).strip()  # Ensure it's a string and remove extra spaces
    if text.isnumeric() or "/" in text:  # Skip pure numbers & timestamps
        return text  
    return text

# Function to translate a batch of texts
def batch_translate_text(texts):
    try:
        texts = [clean_text(t) for t in texts if t]  # Clean & remove empty strings
        if not texts:
            return texts  # Return if all were empty
        
        results = client.translate(texts, target_language="en")  # Batch request
        return [res.get("translatedText", t) for res, t in zip(results, texts)]  # Keep original if translation fails
    except Exception as e:
        print(f"⚠️ Error in batch translation: {e}")
        return texts  # Return original texts if API fails

# Function to process CSV in smaller chunks
def process_csv(input_file, output_file, chunk_size=50):  
    df = pd.read_csv(input_file, encoding="utf-8", dtype=str)  # Read CSV, forcing all data as strings
    text_columns = df.select_dtypes(include=["object"]).columns  # Get text columns

    for column in text_columns:
        print(f"Translating column: {column}...")

        # Detect languages
        detected_languages = df[column].apply(lambda x: detect_language(clean_text(x)) if pd.notna(x) else "unknown")

        # Identify non-English text
        to_translate = df[column][detected_languages != "en"]

        # Process in smaller batches for efficiency
        translated_batches = []
        for i in range(0, len(to_translate), chunk_size):
            batch = to_translate.iloc[i : i + chunk_size].tolist()
            translated_batches.extend(batch_translate_text(batch))
            time.sleep(0.5)  # Short pause to prevent rate limits

        # Replace translated text in the DataFrame
        df.loc[to_translate.index, column] = translated_batches

    # Save translated file
    df.to_csv(output_file, index=False, encoding="utf-8")
    print(f"✅ Translation completed! Saved to {output_file}")

# Define CSV files for translation
files_to_translate = {
    "/Users/sreeharsha/Documents/TGH Data Management Cleaning/Enrollment Survey/Learner Enrollment SPANISH (3.8.23).csv": "translated_spanish.csv",
    "/Users/sreeharsha/Documents/TGH Data Management Cleaning/Enrollment Survey/Learner Enrollment HAITIAN CREOLE (3.8.23).csv": "translated_haitian_creole.csv",
    "/Users/sreeharsha/Documents/TGH Data Management Cleaning/Enrollment Survey/Learner Enrollment CHINESE (3.8.23).csv": "translated_chinese.csv"
}

# Process each CSV file
for input_file, output_file in files_to_translate.items():
    print(f"Processing {input_file}...")
    process_csv(input_file, output_file, chunk_size=50)

print("🎉 All files have been translated successfully!")
```

```python
import pandas as pd
import langdetect
import re
import html
from tqdm import tqdm

# Function to clean text by removing non-English content
def keep_only_english(text):
    try:
        if pd.isna(text) or text.strip() == "":
            return text  # Keep empty values

        # Unescape HTML entities (e.g., &#39; → ')
        text = html.unescape(str(text))  
        # Remove special non-alphanumeric characters
        text = re.sub(r"[^\w\s\$%,.-]", "", text)  
        
        # Split by " - " to separate different languages in the text
        words = text.split(" - ")
        
        # Filter out only English content
        english_words = [word for word in words if langdetect.detect(word) == "en"]
        
        # Join and return English words, if any
        return " - ".join(english_words) if english_words else text  
    except:
        return text  # Return original if detection fails

# Function to replace "Wi" and "Non" with "Yes" and "No" when isolated
def replace_yes_no(text):
    if isinstance(text, str):
        # Replace standalone "Wi" with "Yes" and "Non" with "No"
        text = re.sub(r"(?<!\w)Wi(?!\w)", "Yes", text)  # Only replace if isolated
        text = re.sub(r"(?<!\w)Non(?!\w)", "No", text)  # Only replace if isolated
    return text

# Load the CSV file
input_file = "/Users/sreeharsha/Documents/TGH Data Management Cleaning/Enrollment Survey/translated_haitian_creole.csv"
output_file = "/Users/sreeharsha/Documents/TGH Data Management Cleaning/Enrollment Survey/cleaned_translated_haitian_creole.csv"

df = pd.read_csv(input_file, encoding="utf-8", dtype=str)

# Store original column names
original_columns = df.columns.tolist()

# Create a mapping of original column names to cleaned names
column_mapping = {col: keep_only_english(col) for col in original_columns}

# Rename columns
df.rename(columns=column_mapping, inplace=True)

# Apply cleaning to text columns using the updated column names
for column in tqdm(df.columns, desc="Cleaning columns"):
    if pd.api.types.is_object_dtype(df[column]):  # ✅ Fixed column type check
        df[column] = df[column].apply(keep_only_english)  # Remove non-English text
        df[column] = df[column].apply(replace_yes_no)  # Replace Wi/Non with Yes/No

# Save cleaned file
df.to_csv(output_file, index=False, encoding="utf-8")

print(f"✅ Cleaning complete! Saved to {output_file}")
```

```python
#translating un-processed data

import pandas as pd
from google.cloud import translate_v2 as translate
import os
import re
import tkinter as tk
from tkinter import filedialog
from tqdm import tqdm

# Set up Google Cloud authentication
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "/Users/sreeharsha/Documents/TGH Data Management Cleaning/nth-infusion-452016-j7-9b0ca2e90993.json"

# Initialize Google Translate API client
client = translate.Client()

def translate_texts(texts, batch_size=100):
    """
    Translate a list of texts to English in batches.
    """
    results = []
    for i in range(0, len(texts), batch_size):
        batch = texts[i:i+batch_size]
        try:
            translations = client.translate(batch, target_language="en")
            # Ensure translations is a list (single string may return a dict)
            if isinstance(translations, dict):
                results.append(translations["translatedText"])
            else:
                results.extend([res["translatedText"] for res in translations])
        except Exception as e:
            print(f"⚠️ Error translating batch: {e}")
            results.extend(batch)  # Fallback: return original texts on error
    return results

def replace_yeah(text):
    """
    Replace isolated 'Yeah' with 'Yes' in a given text.
    """
    if isinstance(text, str):
        return re.sub(r"(?<!\w)Yeah(?!\w)", "Yes", text)
    return text

# Create a file dialog to select CSV files
root = tk.Tk()
root.withdraw()
file_paths = filedialog.askopenfilenames(
    title="Select CSV Files for Translation",
    filetypes=[("CSV Files", "*.csv")]
)

if not file_paths:
    print("No files selected. Exiting.")
    exit()

print(f"✅ Selected {len(file_paths)} file(s) for translation.")

# Process each file
for file_path in file_paths:
    print(f"\n🔄 Processing {file_path}...")
    
    try:
        df = pd.read_csv(file_path, encoding="utf-8", dtype=str)
    except Exception as e:
        print(f"⚠️ Error reading file {file_path}: {e}")
        continue

    # Translate column headers
    original_headers = df.columns.tolist()
    print("📢 Translating headers...")
    translated_headers = translate_texts(original_headers)
    # Apply the 'Yeah' replacement to headers as well
    df.columns = [replace_yeah(header) for header in translated_headers]

    # Translate all cell data in columns of type object (i.e. string data)
    for col in tqdm(df.columns, desc=f"Translating data in {file_path}"):
        if df[col].dtype == object:
            non_null_mask = df[col].notnull()
            texts = df.loc[non_null_mask, col].tolist()
            if texts:
                translated_texts = translate_texts(texts)
                # Apply replacement on the translated texts
                translated_texts = [replace_yeah(text) for text in translated_texts]
                df.loc[non_null_mask, col] = translated_texts

    try:
        df.to_csv(file_path, index=False, encoding="utf-8")
        print(f"✅ Updated file saved: {file_path}")
    except Exception as e:
        print(f"⚠️ Error saving file {file_path}: {e}")

print("\n🎉 All selected files have been updated successfully!")

```
